```python
def gaus(x,mu,sigma):
    return stats.norm.pdf(x,mu,sigma)
xes = np.linspace(-400,400,2000)
truck = gaus(xes,0,100)
boat = gaus(xes,130,50)
```

```python
plt.plot(xes,(truck+boat),label='truck+boat')
plt.plot(xes,truck,label='truck')
plt.plot(xes,boat,  label='boat')
plt.xlabel(r'$\Delta t$')
plt.ylabel('probability')
plt.legend()

plt.savefig('Figures/boat.png')
```
