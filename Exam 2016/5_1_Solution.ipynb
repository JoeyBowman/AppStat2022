```python
data = np.genfromtxt('data_lukeLightningLights.txt')
x = data[:,0]
y = data[:,2]
yerr = data[:,3]

first_12_months = x < 12
x_12 = x[first_12_months]
y_12 = y[first_12_months]
yerr_12 = yerr[first_12_months]
```

```python
#Make a plot of the data and the mean of the first 12 months with a residual plot underneath.
fig_12, ax_12 = plt.subplots(2,1, sharex=True, gridspec_kw={'height_ratios': [3, 1]})

ax_12[0].errorbar(x_12, y_12, yerr_12, fmt='o', label='Data')
ax_12[0].hlines(np.mean(y_12),x[0]-1,x_12[-1]+1, color='k', linestyle='--')
res_12 = y_12-np.mean(y_12)
std_res_12 = np.std(res_12)
ax_12[1].errorbar(x_12,res_12,np.ones(len(res_12))*std_res_12, fmt='o', label='Residuals')
ax_12[1].hlines(0,x[0]-1,x_12[-1]+1, color='k', linestyle='--')
ax_12[1].set_xlabel('Time [months]')
```
